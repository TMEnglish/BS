```python
def print_relative_errors(deviate, target, transform=None):
    errors = relative_error(deviate, target)
    if not (transform is None):
        errors = transform(errors)
    print('   Min :', float(np.min(errors)))
    print('   Mean:', float(np.mean(errors)))
    print('   Max :', float(np.max(errors)))
```

```python
def gamma_mixture(n_classes, delta='5e-4', weight='1e-3', alpha='0.5',
                  beta='500', approx=False):
    delta = mp_float(delta)
    weight = mp_float(weight)
    if approx:
        centerpoints = delta * np.arange(1, n_classes)
        tail_masses = delta * gamma_pdf(centerpoints, alpha, beta)
        mass_at_zero = 1 - mp.fsum(tail_masses)
    else:
        endpoints = delta * np.arange(1, n_classes + 1) - delta / 2
        cumulative = gamma_ccdf(endpoints, alpha, beta)
        tail_masses = cumulative[:-1] - cumulative[1:]
        mass_at_zero = 1 - cumulative[0]
    upper_tail = weight * tail_masses
    lower_tail = (1 - weight) * tail_masses[::-1]
    p = np.concatenate((lower_tail, [mass_at_zero], upper_tail))
    p /= mp.fsum(p)
    return p


def basenerize(p):
    """
    Set middle element of array `p` to the immediately preceding element. 
    """
    n = len(p) // 2
    p[n] = p[n-1]
    return p


def f_matrix(p, normed=True):
    """
    Returns [f_ij] matrix; columns are normalized if `normed` is true.
    """
    n = len(p) // 2 + 1
    f = np.empty((n, n), dtype=type(p[0]))
    for j in range(n):
        col = f[:,n-j-1]
        col[:] = p[j:n+j] 
        if normed:
            col /= math.fsum(col)
    return f


class Derivative(object):
    """
    Wrap the W matrix of Basener and Sanford Section 4.
    
    An object d in this class is called d(t, s), where s is the state
    of the system at time t, to obtain the derivative at time t.
    However, the time t is ignored. The derivative is the product of
    the n-by-n matrix W and the n-by-1 vector s.
    """
    def __init__(self, f_matrix, factors):
        """
        Prepare the W matrix of Basener and Sanford Section 4.
        """
        n = len(factors.birth)
        self.factors = factors
        #
        # Columns in the `f_matrix` are scaled by corresponding
        # birth factors. Then the death factor is subtracted from
        # elements on the main diagonal.
        self.W = f_matrix * factors.birth
        self.W[np.diag_indices(n)] -= factors.death
    
    def __call__(self, ignored_time, state):
        """
        Returns vector of derivatives at `state`.
        """
        # Multiply n-by-n W by n-by-1 state.
        return np.dot(self.W, state)
    
    def alt(self, state, include):
        d = np.zeros_like(state)
        np.dot(self.W[include, include], state[include], out=d[include])
        return d
```

## Replicating the defective simulation

```python
def extended_euler(derivative, initial_freqs, n_years,
                   steps_per_year=2**10, zeroing='none'):
    """
    Uses the Euler forward method to solve the initial value problem
    when `zeroing` is 'none'.
    
    To replicate the results of B&S, set `steps_per_year` to 1 and
    `zeroing` to 'incorrect'.
    """
    # Allocate storage for solutions in years 0, 1, ..., `n_years`, and
    # set the solution in year 0 to the given initial frequencies.
    solution = np.empty((n_years + 1, len(initial_freqs)))
    solution[0] = initial_freqs
    #
    if zeroing == 'correct':
        # Zero initial solutions that are small in relation to the sum
        # of the initial solutions. If the i-th element of `zeroes` 
        # is true, then all subsequent solutions for class i are zero.
        zeroes = solution[0] < 1e-9 * np.sum(solution[0])
        solution[0, zeroes] = 0
    elif zeroing == 'incorrect':
        # Erroneously omit the initial zeroing, as Basener does.
        pass
    #
    # The 1-D array `s` always contains the latest solutions.
    s = np.array(solution[0])
    #
    for year in range(1, n_years + 1):
        #
        # In each year, there are `steps_per_year` integration steps.
        for _ in range(steps_per_year):
            #
            # In each step, the solutions in `s` are updated. The step
            # size is the reciprocal of the number of steps per year.
            s += 1 / steps_per_year * derivative(None, s)
            #
            if zeroing == 'correct':
                # The present solution for class i is zeroed if a 
                # previous solution for class i is zero, or the present
                # solution is small in relation to the sum of solutions.
                relatively_small = s < 1e-9 * math.fsum(s)
                zeroes = np.logical_or(relatively_small, zeroes)
                s[zeroes] = 0
            elif zeroing == 'incorrect':
                # Zero solutions that are small in relation to the
                # maximum of the solutions, as Basener does.
                relatively_small = s < 1e-9 * np.max(s)
                s[relatively_small] = 0
        #
        # Scale solutions to keep the maximum of their exponents at zero.
        # Scaling by an integer power of two maintains full precision,
        # because the mantissas of the solutions are unchanged.
        s *= 2 ** -round(math.log(np.max(s), 2))
        #
        # Save the year-end solutions.
        solution[year] = s
    #
    return solution


def runge_kutta(derivative, initial_freqs, n_years, max_step_size=1/128):
    times = np.linspace(0, n_years, n_years + 1)
    return ivp_solution(derivative, initial_freqs, times, max_step_size)
```

```python
def run(n_years, basenerized=True, **args):
    n_classes = len(bs5_4['m'])
    p = gamma_mixture(n_classes, approx=True)
    if basenerized:
        basenerize(p)
    f = f_matrix(convert(p, float), normed=False)
    factors = Factors(n_classes, exclude_max=True)
    derivative = Derivative(f, factors)
    initial_freqs = bs5_4['P']
    initial_freqs /= math.fsum(initial_freqs)
    return extended_euler(derivative, initial_freqs, n_years, **args)

n_years = len(bs5_4['Psolution']) - 1
replicate = run(n_years, basenerized=True, steps_per_year=1,
                zeroing='incorrect')
rs = replicate[-1]
bs = np.array(bs5_4['Psolution'][-1])
rs /= math.fsum(rs)
bs /= math.fsum(bs)

lengths_match = len(rs[rs > 0]) == len(bs[bs > 0])
if lengths_match:
    zeros_match = np.all(np.equal(np.where(rs > 0), np.where(bs > 0)))
else:
    zeros_match = False

print('Comparing B&S solutions to ours in the last year...')
print('Identical numbers of zero solutions:', lengths_match)
print('The locations of zero solutions match:', zeros_match)
if zeros_match:
    print('Maximum absolute relative error:',
          maximum_absolute_relative_error(bs[bs > 0], rs[rs > 0]))
    print("Errors of B&S's nonzero solutions relative to ours:")
    print_relative_errors(bs[bs > 0], rs[rs > 0])
```

```python
# %load ./Code/bs_plot.py
def bs_plot(frequencies, m, intermediate, ls=['-', '-', '-']):
    fig, ax = plt.subplots()
    ax.set_xlabel('Fitness')
    ax.set_ylabel('Relative Frequency')
    years = [0, intermediate, len(frequencies) - 1]
    colors = ['r', 'b', 'limegreen']
    #for year, color, style in zip(years, colors, ls):
    for year, style in zip(years, ls):
        p = frequencies[year] / math.fsum(frequencies[year])
        mean = math.fsum(p * m)
        label = 'Year {0} (mean {1:.3f})'.format(year, mean)
        #ax.plot(m[p>0], p[p>0], color=color, label=label, ls=style)
        ax.plot(m[p>0], p[p>0], label=label, ls=style)
    ax.legend(loc='best')
    return fig, ax
```

## Direct simulation of alternative model

```python
def alt_simulation(derivative, initial_freqs, n_years, 
                   steps_per_year=2**10):
    """
    Restricts simulation to classes with positive frequencies.
    """
    # Allocate storage for solutions in years 0, 1, ..., `n_years`, and
    # set the solution in year 0 to the given initial frequencies.
    solution = np.empty((n_years + 1, len(initial_freqs)))
    solution[0] = initial_freqs
    solution[0, initial_freqs < 1e-9] = 0
    #
    # Set `include` to a slice that includes only positive frequencies.
    include = slice_to_support(solution[0])
    #
    # The 1-D array `s` always contains the latest solutions. The view
    # `p` of `s` contains all and only positive solutions.
    s = np.array(solution[0])
    p = s[include]
    #
    for year in range(1, n_years + 1):
        #
        # In each year, there are `steps_per_year` integration steps.
        for _ in range(steps_per_year):
            #
            # In each step, the solutions in `s` are updated. The step
            # size is the reciprocal of the number of steps per year.
            # The derivative is zero for solutions that are zero.
            s += 1 / steps_per_year * derivative.alt(s, include)
            #
            # Make the `include` slice narrower if any of the positive
            # solutions has fallen below threshold.
            p[p < 1e-9 * math.fsum(p)] = 0
            if p[0] == 0 or p[-1] == 0:
                include = slice_to_support(s)
                p = s[include]
        #
        # Scale solutions to keep the maximum of their exponents at zero.
        # Scaling by an integer power of two maintains full precision,
        # because the mantissas of the solutions are unchanged.
        s *= 2 ** -round(math.log(np.max(s[include]), 2))
        #
        # Save the year-end solutions.
        solution[year] = s
    #
    return solution
```

```python
%%time

n_classes = len(bs5_4['m'])
p = gamma_mixture(n_classes, approx=True)
f = f_matrix(convert(p, float), normed=False)
factors = Factors(n_classes, exclude_max=True)
derivative = Derivative(f, factors)
initial_freqs = bs5_4['P']
initial_freqs /= math.fsum(initial_freqs)
alt_results = alt_simulation(derivative, initial_freqs, 5000, 128)
```
