<h1>Proof of unbounded increase of algorithmic specified complexity…</h1>

<strong>… without infinity and the empty string.</strong>

Some readers have objected to my simple proof that computable transformation $f(x)$ of a binary string $x$ can result in an infinite increase of algorithmic specified complexity (ASC). Here I give a less-simple proof that there is no upper bound on the difference in ASC of $f(x)$ and $x.$ To put it more correctly, I show that the difference can be any positive real number.

<strong>Definitions.</strong> Let $\mathcal{X} := \{0, 1\}^*\!,$ the set of all binary strings. For all $x$ in $\mathcal{X},$ the <em>algorithmic specified complexity</em> of $x$ is defined

$$A(x; c, p) := -\!\log_2 p(x) - K(x|c),$$

where $c$ is an element of $\mathcal{X},$ and $p$ is a distribution of probability over $\mathcal{X}.$ Here $K(x|c)$ is the algorithmic complexity of $x$ conditioned on $c.$

<strong>Theorem.</strong> Let $f(x) = y$ be a constant function on $\mathcal{X}.$ Let $x$ and $c$ be elements of $\mathcal{X}$ for which $K(x|c) > K(y|c)$ holds. Also let $\alpha$ be a positive real number. There exists a distribution $p$ of probability over $\mathcal{X}$ such that

$$A(f(x); c, p) - A(x; c, p) = \alpha.$$

<strong>Proof.</strong> Noting that $f(x) = y$ by hypothesis,

\begin{align*}
   A(f(x); c, p) - A(x; c, p)
   &= A(y; c, p) - A(x; c, p) \\
   &= -\!\log_2 p(y) - K(y|c) - [-\!\log_2 p(x) - K(x|c)] \\
   &= \log_2 p(x) - \log_2 p(y) + K(x|c) - K(y|c) \\
   &= \log_2 \frac{p(x)}{p(y)} + k,
\end{align*}

where $k = K(x|c) - K(y|c) > 0$ and $x \neq y.$ In the case of $\alpha \geq k,$ set $p(x) = .5,$ and solve for $p(y) \leq .5$ in terms of $\alpha$:

\begin{align*}
   \alpha &= \log_2 \frac{p(x)}{p(y)} + k \\
   \alpha - k &= \log_2 \frac{.5}{p(y)} \\
   2^{\alpha - k} &= \frac{.5}{p(y)} \\
   p(y) &= \frac{2^{-1}}{2^{\alpha - k}} \\
   p(y) &= 2^{k- \alpha - 1}.
\end{align*}

Observe that $p(y) = 2^{k- \alpha - 1}$ is precisely equal to $.5$ if and only if $\alpha = k,$ and is strictly less than $.5$ when $\alpha > k.$ Thus $p$ distributes probability mass $1 - [p(x) + p(y)] \geq 0$ over the nonempty set $\mathcal{X} \backslash \{x, y\}$ excluding $x$ and $y.$ In the case of $\alpha < k,$ set $p(y) = .5,$ and similarly solve for $p(x) < .5$ in terms of $\alpha.$ <em>QED.</em>

is a <em>canonical specified complexity model</em> if and only if the <em>scaling constant</em> $r \geq \nu(\mathcal{X}).$ In the special case of $r = \nu(\mathcal{X}),$ which Montañez characterizes as ideal, the expected specified complexity of $X$ is

\begin{align*}
   E \!\left[ SC(X) \right] 
      &= E\!\left[ -\!\log_2 r  \frac{p(X)}{\nu(X)} \right] \\
      &= -E\!\left[ \log_2 \frac{p(X)}{\nu(X)/r} \right] \\
      &= -E\!\left[ \log_2 \frac{p(X)}{ \nu(X) / \nu(\mathcal{X})} \right] \\
      &= -E\!\left[ \log_2 \frac{p(X)}{\nu^*(X)} \right] \\
      &= -D(\nu^*||p),
\end{align*}

where $D(\nu^*||p)$ is the Kullback-Leibler divergence of $\nu^*$ from $p.$ Note that $D(q||p) \geq 0$ for all distributions $q$ of probability over $\mathcal{X},$ with equality only in the case of $q = p.$ Returning to the general case of $r \geq \nu(\mathcal{X}),$ we define a regularizing constant $\rho = \nu(\mathcal{X}) / r.$ That is, $\rho \, r = \nu(\mathcal{X})$ and $\rho \leq 1.$ Continuing from the second step of the derivation above,

\begin{align*}
   -E\!\left[ \log_2 \frac{p(X)}{\nu(X)/r} \right] 
      &= -E\!\left[ \log_2 \frac{p(X)/\rho}{\nu(X)/(\rho \, r)} \right] \\
      &= -E\!\left[ \log_2 \frac{p(X)}{\nu(X)/\nu(\mathcal{X})} - \log_2 \rho \right] \\
      &= -E\!\left[ \log_2 \frac{p(X)}{\nu^*(X)} \right] - E \!\left[ -\!\log_2 \rho \right] \\
      &= -D(\nu^*||p) + \log_2 \rho.
\end{align*}

That is, the expected specified complexity is, up to an additive constant depending on the choice of scaling constant in the canonical specified complexity "model," the negative Kullback-Leibler divergence of the regularized specification function from the probability distribution function. (What is modeled by this "model," Montañez does not say, and I cannot guess. I would say that he has defined a canonical form of specified complexity functions. Let $X$ be a random variable with distribution $p$ of probability over the set $\mathcal{X} := \{ 0, 1\}^*$ of binary strings, and let $C$ be an element of $\mathcal{X}.$ (Confusingly, Nemati and Holloway use the calligraphic $\mathcal{X}$ to denote both the set and the random variable. It is essential, in the following, to understand that the random variable $X$ takes values in the set $\mathcal{X},$ i.e., the set of all binary strings.) Define probability distribution

$$\mathbf{m}_C^*(x) := \frac{\mathbf{m}_{/C}(x)}{\mathbf{m}_{/C}(\mathcal{X})},$$

for all $x$ in $\mathcal{X},$ where $\mathbf{m}_{/C}(x) := 2^{-K(x|C)}$ as in Equation 14 of Nemati and Holloway, and

$$\mathbf{m}_{/C}(\mathcal{X}) := \sum_{z \in \mathcal{X}} \mathbf{m}_{/C}(z).$$

It follows that

$$K(x|C) = -\!\log_2 \mathbf{m}_C^*(x) - \log_2 \mathbf{m}_{/C}(\mathcal{X}),$$

and

\begin{align*}
E[ASC(X, C, p)]
&amp;= E[-\!\log_2 p(X) - K(X|C)] \\
&amp;= E[-\!\log_2 p(X) + \log_2 \mathbf{m}_C^*(X) + \log_2 \mathbf{m}_{/C}(\mathcal{X})] \\
&amp;= E\left[\log_2 \frac{\mathbf{m}_C^*(X)}{p(X)}\right] + \log_2 \mathbf{m}_{/C}(\mathcal{X}) \\
&amp;= -D(p||\mathbf{m}_C^*) + \log_2 \mathbf{m}_{/C}(\mathcal{X}).
\end{align*}

That is, the expected value of $ASC(X, C, p)$ is, up to an additive constant, the negative Kullback-Leibler divergence of $\mathbf{m}_C^*$ from $p.$ A property of the Kullback-Leibler divergence is that $D(p||\mathbf{m}_C^*) \geq 0,$ with equality if and only if $p = \mathbf{m}_C^*.$ By Equation 15 of Nemati and Holloway, $\mathbf{m}_{/C}(\mathcal{X}) &lt; 1.$ Thus $\log_2 \mathbf{m}_{/C}(\mathcal{X}),$ which depends only on the binary string $C$ and the universal prefix machine $U$ in terms of which $K(x|C) \equiv K_U(x|C)$ is defined, is negative.
