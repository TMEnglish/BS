```python
def binned_gamma_mixture(n_bins, bin_width='5e-4', weight='1e-3',
                         alpha='0.5', beta='500', normed=True):
    """
    Returns binned mixture of a Gamma distribution and its reflection.
    
    The middle of the `n_bins` bins is centered on zero. The weight of
    the Gamma(`alpha`, `beta`) distribution in the mixture is `weight`,
    and the weight of the reflection of the distribution is `1 - weight`.
    If `normed` is true, then the bin probabilities are scaled so that
    they sum to 1.
    """
    # Calculate endpoints of width-w bins centered on points w, 2w, ...,
    # nw, where n is the number of bins in the upper tail.
    n = n_bins // 2
    w = mp_float(bin_width)
    bin_endpoints = w * mp.linspace(1/2, n + 1/2, n + 1)
    #
    # Difference the complementary CDF at the endpoints of the bins to
    # obtain the probability masses of the bins.
    complementary_cdf = gamma_ccdf(bin_endpoints, alpha, beta)
    tail_masses = complementary_cdf[:-1] - complementary_cdf[1:]
    #
    # Weight the masses (reversed for the lower tail) to obtain masses
    # for the bins in the tails of the mixture distribution.
    weight = mp_float(weight)
    upper_tail = weight * tail_masses
    lower_tail = (1 - weight) * tail_masses[::-1]
    #
    # The mass of the bin centered on zero, [-w/2, w/2], is CDF(w/2).
    mass_at_zero = 1 - complementary_cdf[0]
    #
    # Assemble the bin masses into a single array.
    p = np.concatenate((lower_tail, [mass_at_zero], upper_tail))
    #
    if normed:
        # Scale the masses so that they sum to 1.
        p /= mp.fsum(p)
    return p


def mutation_matrix(p):
    """
    Returns 2-D array of mutation rates, denoted f by B&S.
    
    The 1-D array `p` is of length 2n + 1. The probability is `p[i-j+n]`
    that the effect of mutation on fitness is i - j units.
    
    The returned matrix f is n-by-n. For off-diagonal elements, f[i, j]
    is equal to `p[i-j+n]`. The value of f[j, j] is set to make the sum
    of elements of the j-th column equal to 1.
    """
    n = len(p) // 2 + 1
    f = np.empty((n, n), dtype=float)
    for j in range(n):
        col = f[:,n-j-1]
        col[:] = p[j:n+j] 
        col[n-j-1] += 1 - math.fsum(col)
    return f


class Derivative(object):
    """
    Wrap the W matrix of Basener and Sanford (Section 4).
    
    An object d in this class is called d(t, s), where s is the state
    of the system at time t, to obtain the derivative at time t.
    However, the time t is ignored. The derivative is the product of
    the n-by-n matrix W and the n-by-1 vector s.
    """
    def __init__(self, mutation_matrix, factors):
        """
        Prepare the W matrix of Basener and Sanford Section 4.
        """
        n = len(factors.birth)
        self.factors = factors
        #
        # Columns in the mutation matrix are scaled by corresponding
        # birth factors. Then the death factor is subtracted from
        # elements on the main diagonal.
        self.W = mutation_matrix * factors.birth
        self.W[np.diag_indices(n)] -= factors.death
    
    def __call__(self, ignored_time, state, include=None):
        """
        Returns vector of derivatives at `state`.
        
        If a slice `include` is provided, then the calculation of
        derivatives is restricted to `state[include]`. For components
        excluded by the slice, the derivatives are set to zero.
        """
        if include is None:
            # Multiply n-by-n W by n-by-1 state.
            return np.dot(self.W, state)
        #
        d = np.zeros_like(state)
        np.dot(self.W[include, include], state[include], out=d[include])
        return d
```
