## Review of the continuous-time model


Basener and Sanford develop the model in [Section 3](https://link.springer.com/article/10.1007%2Fs00285-017-1190-x#Sec9) of the article. 

###### Without mutation

In the absence of mutation, all offspring are identical in type to their parents, and the instantaneous rate of change in the frequency $P_i$ of type $i$ in the population is  the difference in the birth rate $b_i P_i$ and the death rate $d_i P_i.$ With the Malthusian growth factor, also called fitness, $m_i$ defined as the difference of the birth factor $b_i$ and the death factor $d_i,$

\begin{align*}
\frac{\mathrm{d} P_i}{\mathrm{d}t} 
    &= m_i P_i \tag{3.1} \\
    &= (b_i - d_i) P_i \\
    &= b_i P_i - d_i P_i.
\end{align*}

It is important to understand that the growth *rate* $m_i P_i$ is proportional to the present frequency of the type, and that the growth *factor* $m_i$ is the coefficient of proportionality. Given the frequency $P_{0,i}$ of type $i$ in the population at time 0, we obtain

\begin{equation}
P_{ti} = P_{0,i} \exp(t \cdot m_i)
\end{equation}

as the frequency of type $i$ at all times $t$ by solution of the differential equation (3.1).

###### With mutation

In the presence of mutation, the rate at which organisms of type $i$ are born is the sum of the rates $f_{ij} b_j P_j$ at which parents of all types $j$ give birth to organisms of type $i,$ and

\begin{align*}
\frac{\mathrm{d} P_i}{\mathrm{d}t} 
    &= \sum_j f_{ij} b_j P_j - d_i P_i. \tag{3.2}
\end{align*}

That is, the $j$-th column of the *mutation matrix* $F = [f_{ij}]$ specifies the distribution over types $i$ of the progeny of parents of type $j.$ To put it more directly, $f_{ij}$ is the fraction of births to parents of type $j$ that are of type $i.$ If $F$ is the identity matrix, then there is no mutation (offspring of type $i$ come only from parents of type $i$), and [Equation 3.2](https://link.springer.com/article/10.1007%2Fs00285-017-1190-x#Equ8) reduces to [Equation 3.1](https://link.springer.com/article/10.1007%2Fs00285-017-1190-x#Equ7).

## Theorem 2

[Theorem 2](https://link.springer.com/article/10.1007%2Fs00285-017-1190-x#FPar2)


\begin{equation*}
B_i^\mathrm{in} = \sum_j f_{ij} B_j^\mathrm{out}.
\end{equation*}

That is, $f_{ij}$ is the fraction of the births to type $j$ parents that are of type $i.$

Reorganizing the definitions of Basener and Sanford, we write the instantaneous rate of change in the frequency $P_i$ of type $i$ within the population as

\begin{equation*}
\frac{\mathrm{d} P_i}{\mathrm{d}t} = B_i^\mathrm{in} - d_i P_i,
\end{equation*}

where $d_i P_i$ is the rate at which organisms of type $i$ die,

\begin{equation*}
B_i^\mathrm{in} = \sum_j f_{ij} B_j^\mathrm{out}
\end{equation*}

is the rate at which organisms of type $i$ are born,


\begin{equation*}
B_j^\mathrm{out} = b_j P_j
\end{equation*}

is the rate at which organisms of type $j$ give birth, and $f_{ij}$ is the fraction of the births to organisms of type $j$ that are of type $i.$ That is, the $j$-th column of the matrix $[f_{ij}]$ specifies the distribution of the progeny of type-$j$ organisms over the various types $i.$ The Malthusian growth factor of the $i$-th type, sometimes referred to as its *fitness*, is $m_i = b_i - d_i.$ (Note that I have reorganized the definitions of Basener and Sanford.) 

Theorem 2 states that the instantaneous rate of change in $\bar{m},$ the mean fitness for the population, is

\begin{equation*}
\frac{\mathrm{d}\bar{m}}{\mathrm{d}t} 
   = \mathrm{Var}(m) + \frac{1}{\sum_i P_i} \sum_i (B_i^\mathrm{in} - B_i^\mathrm{out}) (m_i - \bar{m}), \tag{3.4}
\end{equation*}

where $\text{Var}(m)$ is the variance of the population in fitness.

## Review of the discrete-time computation



## Application of Theorem 2 to a discrete-time system

The matrix $[b_j\, f_{ij}]$ is available as the `birthing` attribute of a `Population` instance. The quantities $B_i^\mathrm{out}$ are obtained by taking the pointwise product of the `Population` frequencies and the birth factors. Use of effective growth factors and effective birth factors is important&hellip;

```python
def theory(initial, mutations, m):
    size = initial.norm()
    p = initial[:] / size
    mean = mp.fsum(m * p)
    variance = mp.fsum(m ** 2 * p) - mean ** 2
    b = m - m[0]
    b_in = np.dot(mutations.matrix() * b, initial)
    b_out = initial * b
    return variance + mp.fsum((b_in - b_out) * (m - mean)) / size

def actual(pop, effective=True):
    mean = pop.mean(effective)
    pop.update()
    return pop.updates_per_year * (pop.mean(effective) - mean)
```

The effective growth factors for the computational process are $m^\prime_i = \ln(m_i + 1).$ That is, in the absence of mutation, the number of organisms of type $i$ after $t$ annual updates of the frequency distribution is 

$$P_{ti} = P_{0,i} \exp(t \cdot m^\prime_i).$$

The question is how to define the effective birth rates $b^\prime_i$ consistently when there is mutation. It's not clear that things will pan out under the assumption that $m^\prime_i = b^\prime_i - d^\prime.$ But let's start with the assumption, and consider the consequences. It *seems* that $b^\prime_0 = b_0 = 0$ is the only biologically sensible choice, with the implication that $d^\prime = -m^\prime_0.$ Then $b^\prime_i = m^\prime_i - m^\prime_0.$


Then, in the case that there exists a $k$ for which $m^\prime_k = 0,$ ??? that $b^\prime_k = d.$ 

\begin{align*}
P_{ti} 
   &= P_{0,i} \exp(t \cdot m^\prime_i) \\
   &= P_{0,i} \exp(t \cdot (b^\prime_i - d^\prime)) \\
   &= P_{0,i} \exp(t \cdot b^\prime_i - t \cdot d^\prime) \\
   &= P_{0,i} \frac{\exp(t \cdot b^\prime_i) }{ \exp(t \cdot d^\prime)} \\
\end{align*}

Note that we're actually checking both my software and their theorem. If the check fails, then something is wrong with one or both of the software and Equation 3.4. If the check succeeds, then the result increases our confidence in both the software and the theorem.

Our first case is a unit-size population with mean fitness of zero, i.e., with $\sum_i P_i = 1$ and 

$$\bar{m} = \sum_i m_i P_i = 0.$$

As in the numerical studies of Section 5.4, the death factor $b_i - m_i = d$ for all types $i.$ Then 

\begin{align*}
\mathrm{Var}(m) 
   &= \sum_i m_i^2 P_i - \bar{m}^2 \\
   &= \sum_i m_i^2 P_i, \\
\end{align*}

and Equation 3.4 is equivalent to

\begin{align*}
\frac{\mathrm{d}\bar{m}}{\mathrm{d}t} 
   &= \sum_i m_i^2 P_i + \sum_i (B_i^\mathrm{in} - B_i^\mathrm{out}) m_i \\
   &= \sum_i m_i (m_i P_i + B_i^\mathrm{in} - B_i^\mathrm{out}) \\
   &= \sum_i m_i (m_i P_i + B_i^\mathrm{in} - b_i P_i) \\
   &= \sum_i m_i \left[ B_i^\mathrm{in} + (m_i - b_i) P_i \right] \\
   &= \sum_i m_i ( B_i^\mathrm{in} - d P_i ) \\
   &= \sum_i m_i B_i^\mathrm{in} - d \sum_i m_i P_i \\
   &= \sum_i m_i B_i^\mathrm{in} - d \cdot \bar{m} \\
   &= \sum_i m_i B_i^\mathrm{in} . \\
\end{align*}

To simplify further, suppose that all of the population is of the type $k$ with growth factor $m_k = 0$ (and birth factor $b_k = d$), which is to say that $P_k = 1$ and $P_i = 0$ for all $i \neq k.$ Then

\begin{align*}
\frac{\mathrm{d}\bar{m}}{\mathrm{d}t} 
   &= \sum_i m_i B_i^\mathrm{in}  \\
   &= \sum_i m_i \sum_j b_j\, f_{ij} P_j \\
   &= \sum_i m_i b_k\, f_{ik} \\
   &= b_k \sum_i m_i\, f_{ik} \\
   &= d \sum_i m_i\, f_{ik} .
\end{align*}
