### Distribution of mutation effects

According to the introduction to Section 5 of the article:

>The distribution can be modeled using a Gamma probability distribution (Kimura 1979).

>$$f(s^\prime) = \alpha^\beta e^{-\alpha s^\prime} {s^\prime}^{\beta - 1} / \Gamma(\beta),$$

>where $s^\prime$ is the change in fitness (in absolute value), $f(s^\prime)$ is the probability for this change in fitness to occur from parent to child, $\alpha=\beta/\bar{s}^\prime$ where $\bar{s}^\prime$ is the mean selective disadvantage, and   $\beta$  is a parameter, called either the shape parameter or rate parameter, with $0 < \beta \leq 1.$ Kimura suggests that a typical value for $\bar{s}^\prime$ is $10^{-3}$ for deleterious mutations, which is the value we use.

>The parameter $\beta$ determines the tendency for mutations to be neutral; in the limiting case as $\beta\rightarrow 0,$ all mutations are neutral and in the case with $\beta=1,$ we regain Ohata’s [*sic*] model (Ohta 1977) in which there are not a sufficient number of nearly-neutral mutaitons [*sic*], according to Kimura (1979). We choose $\beta=0.5$ which is the estimation given by Kimura (1979).

>$$f(s^\prime)=500^{0.5} e^{-500s^\prime} {s^\prime}^{- 0.5} / \Gamma(0.5) \approx 12.6e^{-500s^\prime} {s^\prime}^ {-0.99}$$

>If we use $\Delta_{i,j}$ to denote the change in fitness from $m_i$ to $m_j$ then our formula becomes

>$$f_{i,j} \approx 12.6e^{-500 \Delta_{i,j}} \Delta_{i,j}^{-0.5} \tag{5.1}$$

>This distribution used by Kimura only accounts for deleterious mutations. Put simply, Basener's script silently kills many of the births, and greatly reduces the incidence of beneficial mutation in those that it does not kill.

Basener and Sanford did not need the botched reweighting to obtain the "collapsing" population of Section 5.4. Here we retain only the gimmick for reducing the probability that mutation has zero effect. The comments in the following cell explain the operations we are about to perform in Python.

````python
# %load ../Code/gamma.py
class Gamma(EffectsDistribution):
    """
    A discretized Gamma distribution and its reflection in the zero-effect axis.
    """
    def __init__(self, factors, alpha=0.5, beta=0.5/0.001, density=True,
                       normed=False):
        """
        Defines a symmetric Gamma distribution with the given parameters.
        
        The `factors` object is an instance of class `Factors`. The Boolean
        parameters `density` and `normed` determine, respectively, whether the
        density function or the cumulative distribution function is used in
        the initial assignment of probability masses to effects (the latter is
        always used for zero effect because the former is undefined at zero),
        and whether the resulting distribution is normalized.
        """
        distribution = stats.gamma(alpha, scale=1/beta)
        super().__init__(factors, distribution, density=density, normed=normed)
        
    def neutral_gimmick(self):
        """
        Assigns the probability of minimally deleterious effect to zero effect.
        """
        self.p[self.zero_index] = self.p[self.zero_index - 1]
    
    def botched_reweight(self, p=0.001):
        """
        Multiplies probabilities of positive effects by `p`, others by 1 - `p`.
        """
        p = float(p)
        self.p[self.zero_index + 1 :] *= p
        self.p[: self.zero_index + 1] *= 1 - p

    def reweight(self, p=mp.mpf('0.001'), botched=False):
        """
        Sets probabilities of positive and non-positive effect to `p` and 1-`p`.
        """
        p = mp.mpf(p) # multiprecision float
        positive = self.p[self.zero_index + 1 :]
        non_positive = self.p[: self.zero_index + 1]
        positive[:] = p / accurate_sum(positive) * positive
        non_positive[:] = (1 - p) / accurate_sum(non_positive) * non_positive
````

**Define the fraction of mutations that are beneficial**

````python
beneficial = mp.mpf('1e-3')
````

**Reflected, reweighted Gamma distribution described by Basener and Sanford**

1. A Gamma distribution over effects is symmetrized by reflection in the zero-effect axis, and is discretized.
2. The probabilities in the positive tail are scaled so that they sum to `beneficial`. That is, if the probabilities of positive effects initially sum to $b,$ then all of them are multiplied by `beneficial` / $b.$
3. The remaining probabilities are scaled similarly, so that they sum to 1 - `beneficial`.

````python
reported = Gamma(factors)
reported.reweight(beneficial)
````

4. The remaining probabilities are scaled so that they sum to 1 - `beneficial`, as above.
